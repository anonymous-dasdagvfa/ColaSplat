```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def plot_four_layers(mask_array):
    """
    mask_array: numpy array of shape (4, H, W)
    画出4个掩码层，1行4列子图
    """
    assert mask_array.shape[0] == 4, "输入必须是4层掩码"

    fig, axes = plt.subplots(1, 4, figsize=(20, 5))
    for i in range(4):
        ax = axes[i]
        im = ax.imshow(mask_array[i], cmap='tab20')
        ax.set_title(f"Segmentation Mask - Layer {i}")
        ax.axis('off')
        fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    plt.tight_layout()
    plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_discrete_masks(s, layer_idx=3):
    layer = s[layer_idx]  # shape (H, W)
    unique_vals = np.unique(layer)

    print(f"Layer {layer_idx} has {len(unique_vals)} unique values: {unique_vals}")

    n = len(unique_vals)
    cols = min(n, 4)
    rows = (n + cols - 1) // cols

    plt.figure(figsize=(4 * cols, 4 * rows))

    for i, val in enumerate(unique_vals):
        mask = (layer == val).astype(np.uint8)

        plt.subplot(rows, cols, i + 1)
        plt.imshow(mask, cmap='gray')
        plt.title(f"Value = {val}")
        plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
s = np.load("/data2/jian/LangSplat/data/lerf_ovs/waldo_kitchen/language_features/frame_00001_s.npy")
print(s.shape)  # 应该是 (C, H, W)
plot_discrete_masks(s, layer_idx=0)
```
